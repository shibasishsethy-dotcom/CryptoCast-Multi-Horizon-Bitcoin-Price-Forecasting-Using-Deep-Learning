```python
def convert_volume(x):
    if pd.isna(x):
        return None
    x = str(x).strip()

    if 'K' in x:
        return float(x.replace('K', '')) * 1_000
    elif 'M' in x:
        return float(x.replace('M', '')) * 1_000_000
    elif 'B' in x:
        return float(x.replace('B', '')) * 1_000_000_000
    else:
        return float(x)

df['Vol.'] = df['Vol.'].apply(convert_volume)
```

# **TRANSFORMER**

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, LayerNormalization, MultiHeadAttention, GlobalAveragePooling1D
from tensorflow.keras.models import Model
```

```python
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout):
    x = MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(inputs, inputs)
    x = Dropout(dropout)(x)
    x = LayerNormalization(epsilon=1e-6)(x + inputs)

    ff = Dense(ff_dim, activation="relu")(x)
    ff = Dropout(dropout)(ff)
    ff = Dense(inputs.shape[-1])(ff)

    return LayerNormalization(epsilon=1e-6)(x + ff)
```

```python
def build_transformer_model(input_shape, output_steps):
    inputs = Input(shape=input_shape)
    x = inputs

    # Transformer blocks
    for _ in range(3):
        x = transformer_encoder(x, head_size=64, num_heads=4, ff_dim=128, dropout=0.1)

    x = GlobalAveragePooling1D()(x)

    x = Dense(64, activation="relu")(x)
    x = Dropout(0.2)(x)

    outputs = Dense(output_steps)(x)  # рҹ”Ҙ key part

    model = Model(inputs, outputs)
    return model
```

```python
transformer_1d = build_transformer_model((60,1), output_steps=1)

transformer_1d.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

history_1d = transformer_1d.fit(
    x_train_1d, y_train_1d,
    validation_split=0.1,
    epochs=50,
    batch_size=16
)
```
